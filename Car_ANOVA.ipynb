## Location

```python
df['Location'].unique()
```

```python
test = [df['Price'][df['Location'] == i] for i in df['Location'].unique()]

f_statistic, p_value = f_oneway(*test)  # 유의수준 0.05 설정 
print("p value : {}".format(p_value.round(2)))

if p_value < 0.05:
    print("Location 집단간의 차이가 가격에 영향을 끼친다.")
else:
    print("Location 집단간의 차이가 가격에 영향을 끼치지 않는다.")
```

## Fuel_Type
